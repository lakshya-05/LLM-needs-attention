```python
import importlib
import tiktoken

print(f"V-{importlib.metadata.version("tiktoken")}")
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
txt = ("Hello, how are you? <|endoftext|> In the sun Withlightbreeze!")
ints = tokenizer.encode(txt, allowed_special={"<|endoftext|>"})
print(ints)
```

```python
ints = tokenizer.encode("Sneyvs Gds")
print(ints)
```

```python
tokenizer.decode(ints)
```

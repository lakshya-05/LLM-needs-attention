```python
class tokenizationV1:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s, i in vocab.items()}

    def encoder(self, text):
        preproc = re.split(r'([,.:;?_!"()\']|--|\s)', txt)
        res = [i.strip() for i in preproc if i.strip()]
        ids = [self.str_to_int[s] for s in res]
        return ids

    def decoder(self, ids):
        txt = " ".join([self.int_to_str[i] for i in ids])
        txt = re.sub(r'\s+([,.?!"()\'])', r'\1', txt)
        return txt
```

```python
tokenizer = tokenizationV1(vocab)
txt = """I had always thought"""

ids = tokenizer.encoder(txt)
print(ids)
```

```python
tokenizer.decoder(ids)
```

```python
class tokenizationV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s, i in vocab.items()}

    def encoder(self, text):
        preproc = re.split(r'([,.:;?_!"()\']|--|\s)', txt)
        res = [i.strip() for i in preproc if i.strip()]
        res = [
            item if item in self.str_to_int
            else "<|unk|>" for item in res
        ]
        ids = [self.str_to_int[s] for s in res]
        return ids

    def decoder(self, ids):
        txt = " ".join([self.int_to_str[i] for i in ids])
        txt = re.sub(r'\s+([,.?!"()\'])', r'\1', txt)
        return txt
        
```

```python
tokenizer = tokenizationV2(vocab)
t1 = "Hello, how are you?"
t2 = "In the rain of oz"

txt  = " <|endoftext|> ".join((t1, t2))
txt
```

```python
ids = tokenizer.encoder(txt)
ids
```

```python
tokenizer.decoder(ids)
```
